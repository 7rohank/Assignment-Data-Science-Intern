```python
from nltk.tokenize import sent_tokenize
import nltk

# Download NLTK tokenizer data
nltk.download('punkt')

# Sample corpus
corpus = [
    "Natural language processing is a subfield of artificial intelligence that focuses on the interaction between computers and humans using natural language.",
    "Machine learning is the study of computer algorithms that improve automatically through experience.",
    "Deep learning is a subset of machine learning, using neural networks with representation learning."
]

# Chunking the text into ~200-300 word chunks
def chunk_text(corpus, chunk_size=200):
    chunks = []
    for text in corpus:
        sentences = sent_tokenize(text)
        current_chunk = []
        current_size = 0
        for sentence in sentences:
            if current_size + len(sentence) <= chunk_size:
                current_chunk.append(sentence)
                current_size += len(sentence)
            else:
                chunks.append(" ".join(current_chunk))
                current_chunk = [sentence]
                current_size = len(sentence)
        if current_chunk:
            chunks.append(" ".join(current_chunk))
    return chunks

# Prepare chunks
chunks = chunk_text(corpus)
print("Chunks:", chunks)
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Load sentence embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for each chunk
embeddings = model.encode(chunks)

# Create FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))

# Save the FAISS index and chunks for later use
faiss.write_index(index, "faiss_index.idx")
with open("chunks.txt", "w") as f:
    for chunk in chunks:
        f.write(chunk + "\n")
```

```python
def retrieve_chunks(query, top_k=3):
    query_embedding = model.encode([query])
    distances, indices = index.search(query_embedding, top_k)
    results = [chunks[idx] for idx in indices[0]]
    return results

# Test retrieval
query = "What is natural language processing?"
print("Retrieved Chunks:", retrieve_chunks(query))
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load a lightweight GPT model (distilgpt2)
tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
model = AutoModelForCausalLM.from_pretrained("distilgpt2")

# Generate an answer using retrieved chunks as context
def generate_answer(query, retrieved_chunks):
    # Combine query and retrieved chunks
    context = " ".join(retrieved_chunks)
    prompt = f"User: {query}\nContext: {context}\nAssistant:"


    inputs = tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True)


    outputs = model.generate(inputs["input_ids"], max_length=200, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)


    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response.split("Assistant:")[-1].strip()

# Test the updated function
retrieved_chunks = retrieve_chunks("What is natural language processing?")
answer = generate_answer("What is natural language processing?", retrieved_chunks)
print("Answer:", answer)
```

```python
from flask import Flask, request, jsonify
import mysql.connector
import datetime

# MySQL config
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'password',
    'database': 'chatbot_db'
}

# Flask app
app = Flask(__name__)

@app.route('/chat', methods=['POST'])
def chat():
    user_query = request.json.get('query')
    if not user_query:
        return jsonify({'error': 'Query is required'}), 400

    # Retrieve and generate response
    retrieved_chunks = retrieve_chunks(user_query)
    answer = generate_answer(user_query, retrieved_chunks)

    # Store history in MySQL
    conn = mysql.connector.connect(**db_config)
    cursor = conn.cursor()
    cursor.execute(
        "INSERT INTO chat_history (role, content, timestamp) VALUES (%s, %s, %s)",
        ('user', user_query, datetime.datetime.now())
    )
    cursor.execute(
        "INSERT INTO chat_history (role, content, timestamp) VALUES (%s, %s, %s)",
        ('system', answer, datetime.datetime.now())
    )
    conn.commit()
    conn.close()

    return jsonify({'answer': answer, 'retrieved_chunks': retrieved_chunks})

@app.route('/history', methods=['GET'])
def history():
    conn = mysql.connector.connect(**db_config)
    cursor = conn.cursor(dictionary=True)
    cursor.execute("SELECT * FROM chat_history ORDER BY timestamp")
    history = cursor.fetchall()
    conn.close()
    return jsonify(history)

if __name__ == '__main__':
    from waitress import serve
    serve(app, host='0.0.0.0', port=5000)

```
